```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

fig.set_facecolor('#333333')
ax.set_facecolor('#333333')

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))

ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Change the color of the x-axis numbers
ax.tick_params(axis='x', colors='white', labelsize=16)

# Change the color of the y-axis numbers
ax.tick_params(axis='y', colors='white', labelsize=16)


x_max = 4
y_max = 4

# Set the axis limits
ax.set_xlim([-1, x_max])
ax.set_ylim([-1, y_max])


# ax.spines['bottom'].set_linewidth(4)
# ax.spines['left'].set_linewidth(4)

plt.rcParams['lines.linewidth'] = 2

arrow_props = dict(arrowstyle="->", color='white', linewidth=3)

ax.annotate("", xy=(x_max, 0), xytext=(-1.1, 0),
            arrowprops=arrow_props)

ax.annotate("", xy=(0, y_max), xytext=(0, -1.1),
            arrowprops=arrow_props)

# increase tick width
ax.tick_params(width=4)



# Draw the vector
origin = [0, 0]  # Origin of the vector
vector = [3.6, 1.8]  # Vector coordinates

# Calculate the vector components
dx = vector[0] - origin[0]
dy = vector[1] - origin[1]


# Draw the vector
ax.quiver(*origin, dx, dy, angles='xy', scale_units='xy', scale=1, color = 'orange', width=0.015)

# Add a label for the vector
vector_label = f"Vector ({vector[0]}, {vector[1]})"
ax.text(vector[0], vector[1], vector_label, fontsize=16, color='white')

ax.text(4.0, 0.2, 'x1' , fontsize=16, color='orange')
ax.text(0.2, 3.8, 'x2' , fontsize=16, color='orange')

ax.vlines(vector[0], 0, vector[1], linestyle="dashed", color='green')
ax.hlines(vector[1], 0, vector[0], linestyle="dashed", color='green')



# Adjust the spacing between the axis and the tick labels
plt.subplots_adjust(left=0.15, bottom=0.15)

plt.savefig('../images/001_axis_and_vector.png', dpi=300, transparent=True, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

fig.set_facecolor('#333333')
ax.set_facecolor('#333333')

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))

ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Change the color of the x-axis numbers
ax.tick_params(axis='x', colors='white', labelsize=16)

# Change the color of the y-axis numbers
ax.tick_params(axis='y', colors='white', labelsize=16)


x_max = 6
y_max = 6

# Set the axis limits
ax.set_xlim([-1, x_max])
ax.set_ylim([-1, y_max])


# ax.spines['bottom'].set_linewidth(4)
# ax.spines['left'].set_linewidth(4)

plt.rcParams['lines.linewidth'] = 2

arrow_props = dict(arrowstyle="->", color='white', linewidth=3)

# drawing x axis
ax.annotate("", xy=(x_max, 0), xytext=(-1.1, 0),
            arrowprops=arrow_props)

# drawing y axis
ax.annotate("", xy=(0, y_max), xytext=(0, -1.1),
            arrowprops=arrow_props)

z_scale = 0.5

z_max = x_max * z_scale
z_min = 1.1 * z_scale

# array of z_tick_position, from (1,1) to (4,4)
z_tick_real_value = [(x, x) for x in range(1, 5)]

z_tick_position = [(x, x*z_scale, y*z_scale) for (x, y) in z_tick_real_value]

# drawing z axis 
ax.annotate("", xy=(2.2, 2.2), xytext=(-0.6, -0.6),
            arrowprops=arrow_props)

for (label, x, y) in z_tick_position:
    ax.text(x, y, '-', fontsize=30, color='white')
    ax.text(x-0.2, y, str(label), fontsize=16, color='white')


# increase tick width
ax.tick_params(width=4)



# Draw the vector
origin = [0, 0]  # Origin of the vector
vector = [3.6, 1.8]  # Vector coordinates

z_value = 2.1

scaled_z_value = z_value * z_scale

# Calculate the vector components
dx = vector[0] - origin[0]
dy = vector[1] - origin[1]


# Draw the vector
ax.quiver(*origin, dx+scaled_z_value, dy+scaled_z_value, angles='xy', scale_units='xy', scale=1, color = 'orange', width=0.015, zorder=3)



ax.vlines(vector[0], 0, vector[1], linestyle="dashed", color='green', zorder=4)
ax.hlines(vector[1], 0, vector[0], linestyle="dashed", color='green', zorder=4)

ax.vlines(vector[0]+scaled_z_value, 0+scaled_z_value, vector[1]+scaled_z_value, linestyle="dashed", color='green')
ax.vlines(scaled_z_value, 0+scaled_z_value, vector[1]+scaled_z_value, linestyle="dashed", color='green')

ax.hlines(vector[1]+scaled_z_value, 0+scaled_z_value, vector[0]+scaled_z_value, linestyle="dashed", color='green')
ax.hlines(scaled_z_value, 0+scaled_z_value, vector[0]+scaled_z_value, linestyle="dashed", color='green')


ax.plot([vector[0], vector[0]+scaled_z_value], [0, 0+scaled_z_value],  linestyle="dashed", color='green')

ax.plot([0, 0+scaled_z_value], [vector[1], vector[1]+scaled_z_value], linestyle="dashed", color='green')

ax.plot([vector[0], vector[0]+scaled_z_value], [vector[1], vector[1]+scaled_z_value], linestyle="dashed", color='green', zorder=20)


# Add a label for the vector
vector_label = f"Vector ({vector[0]}, {vector[1]}, {z_value})"

ax.text(vector[0], vector[1]+1.5, vector_label, fontsize=16, color='white')

ax.text(6.0, 0.2, 'x1' , fontsize=16, color='orange')
ax.text(0.2, 5.8, 'x2' , fontsize=16, color='orange')
ax.text(4.4 * z_scale, 4.4 * z_scale, 'x3' , fontsize=16, color='orange')



# Adjust the spacing between the axis and the tick labels
plt.subplots_adjust(left=0.15, bottom=0.15)

plt.savefig('../images/001_axis_and_3dvector.png', dpi=300, transparent=True, bbox_inches='tight')

# Show the plot
plt.show()
```
