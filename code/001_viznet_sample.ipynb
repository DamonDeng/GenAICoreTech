```python
# import torch for tensor operations
# import matplotlib for plotting

import torch
import matplotlib.pyplot as plt
from viznet import NodeBrush, EdgeBrush, DynamicShow
```

```python
# set font to support Chinese

font_name = 'Songti SC'
plt.rcParams["font.family"] = font_name
```

```python
# sample code to generate image


with DynamicShow((20, 16), '../images/sample_chinese_english.png') as d:

    brush = NodeBrush('nn.input', size='normal')
    node1 = brush >> (1,0)  # paint a node at (x=1, y=0)
    node2 = brush >> (2,0)
    edge = EdgeBrush('->-', lw=2)
    edge >> (node1, node2)  # connect two nodes
    node1.text('中文', 'center', fontsize=10)  # add text to node1
    node2.text('英文', 'center', fontsize=10)
    
    rbox_brush = brush = NodeBrush('box', size=[0.3,0.1], roundness=0.05, color='red')
    
    nodes_list = []
    
    for i in range(3):
        nodes_list.append(rbox_brush >> (i, 2))
        nodes_list[i].text('中文', 'center', fontsize=10)
    
```
